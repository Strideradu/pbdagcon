# Haplotype Consensus Example Using The PB_DAGCon

This IPython Notebook shows how to use the sequence graph alignment library PB_DAGCon to generate consenus sequenece from PacBio raw sequence data. It also shows the long reads are able to catch the variants between two haplotypes. A simple EM-like algorithm can seperate the reads into their haplotype groups using those variants acorss a long distance. With that, we can generate accurate consensus haplotype sequences using a small number of reads without doing any assembly work.  The data set here is modeled after the HLA-A allele with typical single molecule sequencing data. Check the end of the notebook to see the requirement to run this notebook. We suggest to run it in a mulit-core Linux server.

The ``read_node_vector`` function collects the relationship between the alignment nodes and the reads. The information is used for clustering the sequences into different haplotype groups.

```python
def read_node_vector(g, entropy_th = 0.5):
    
    ne, hne = g.get_high_entropy_nodes(coverage_th = 0, entropy_th = entropy_th)
    node_to_entropy = dict( [ (v[1],v[2]) for v in ne ] ) 
    read_ids = set()
    
    for n in g.get_nodes().values():
        for r in n.get_info():
            read_ids.add(r[0]) #r[0]: read id, r[1]: read base
    
    backbone_node_to_reads = {}
    for n in g.get_nodes().values():
        bn = n.get_backbone_node()
        backbone_node_to_reads.setdefault(bn, set() )
        for r_id, r_pos in n.get_info():
            backbone_node_to_reads[bn].add( r_id )
    
    backbone_node_to_pos = g.get_backbone_node_to_pos()
    sn = g.get_sorted_nodes()
    high_entropy_nodes = [ (n, backbone_node_to_pos[n.get_backbone_node()], node_to_entropy[n]) for n in sn if n in node_to_entropy and node_to_entropy[n] > entropy_th]
    
    read_to_nodes = {}
    for r_id in read_ids:
        read_to_nodes[r_id] = [" "] * len(high_entropy_nodes)
        for i,n in enumerate(high_entropy_nodes):
            if r_id in [x[0] for x in n[0].get_info()]:
                read_to_nodes[r_id][i] = n[0].get_base()
            else:
                bs,be = g.get_read_range()[r_id]
                bp = backbone_node_to_pos[n[0].get_backbone_node()]
                if bp >= bs and bp < be:
                    read_to_nodes[r_id][i] = "*"
    

    return read_to_nodes, high_entropy_nodes
```

```python
rv, hen = read_node_vector(g, entropy_th = 0.5)
d = []
for r in rv:
    d.append( (r, "".join(rv[r])) )
d.sort(key=lambda x: x[1])
d.reverse()
```

The code below is a simple EM-like cluster algorithm to cluster the reads and the variants into two groups simultaneously.

```python
import numpy as np

read_to_cluster = {}
node_to_cluster = {}
for  r,l in d:
    read_to_cluster[r] = 0

for  i in range(len(d[0][1])):
    node_to_cluster[i] = 0

for r,l in d:
    if l[0] in ["A","C","G","T"]:
        read_to_cluster[r] = 1
    elif l[0] == "*":
        read_to_cluster[r] = -1

for xx in range(20):
    #update node vector
    node_count = {}
    for r,l in d:
        if read_to_cluster[r] == 1:
            for i,c in enumerate(l):
                if c in ["A", "C", "G", "T"]:
                    node_count[i] = node_count.get(i,0) + 1
                elif c == "*" :
                    node_count[i] = node_count.get(i,0) - 1
        elif read_to_cluster[r] == -1:
            for i,c in enumerate(l):
                if c in ["A", "C", "G", "T"]:
                    node_count[i] = node_count.get(i,0) - 1
                elif c == "*" :
                    node_count[i] = node_count.get(i,0) + 1
    for i, c in node_count.items():
        if c >= 0:
            node_to_cluster[i] = 1
        elif c < 0:
            node_to_cluster[i] = -1
            
    #update read vector
    m = {"A":1, "C":1, "G":1, "T":1, "*":-1, " ":0}
    for r,l in d:
        v = np.array( [ m[c] for c in l ]) 
        u = np.array( [ node_to_cluster[c] for c in range(len(d[0][1])) ])
        if sum(v*u) >= 0:
            read_to_cluster[r] = 1
        elif sum(v*u) < 0:
            read_to_cluster[r] = -1
```

We can pick the variant nodes in the graph and generate a subgraph which only consists of the variant nodes. We can show how many reads support the haplotype-link of the nearby variants. The cell below generates a simple graph data structure for the haplotype information.

```python
G = {}
for r, l in d:
    n1 = -1
    c1 = ""
    for ni, c in enumerate(l):
        if c in ("A", "G", "C", "T"):
            if n1 == -1:
                n1 = ni
                c1 = c
                continue
            else:
                assert hen[n1][0].get_base() == c1
                assert hen[ni][0].get_base() == c
                node1 = (n1, c1,  hen[n1][1], 1 if hen[n1][0].get_is_backbone() else 0)
                node2 = (ni, c, hen[ni][1], 1 if hen[ni][0].get_is_backbone() else 0)
                G.setdefault( (node1, node2), 0 )
                G[ (node1, node2) ] += 1
                n1 = ni
                c1 = c
```

Using ``networkx`` to convert the data structure to a dot file so we can use ``graphviz`` to layout the graph.

```python
import networkx as nx
DG = nx.DiGraph()
u = np.array( [ node_to_cluster[c] for c in range(len(d[0][1])) ])
for n1,n2 in G: 
    v = G[ (n1,n2) ]
    if v < 12: 
        continue
    #print n1[0], u[n1[0]], n1[3], n2[0], u[n2[0]], n2[3], v
    n1 = "%d_%s_%d_%d" % n1
    n2 = "%d_%s_%d_%d" % n2
    DG.add_edge(n1,n2, penwidth=v/50+1, weight=v/20+1, label="%d" % v)
nx.write_dot(DG, "phasing.dot" )
import os
os.system("dot phasing.dot -Tpng -s24 -x -ophasing.png")
```
